```python
#%matplotlib inline
import gc
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import matplotlib.colors as colors

def cm2inch(value):
    return value/2.54
```

```python
tsvdata = "../DATA/pandas_outdegree_out/yearly_count_raw_ALL.tsv"
countdata = pd.read_csv(tsvdata, delimiter ="\t", index_col=0)
```

```python
countdata[0:10]
```

```python
tsvdata_norm = "../DATA/pandas_outdegree_out/yearly_count_normalized_ALL.tsv"
countdata_norm = pd.read_csv(tsvdata_norm, delimiter ="\t", index_col=0)
```

```python
gc.collect()
```

```python
YearBoundary = 21

fig = plt.figure(figsize =(cm2inch(19)*2,cm2inch(8)*2))

pub = 0
cur_pub_df = countdata[(countdata['PubYear'] == pub) & (countdata['CiteCount'] != 0)]
temparray = []
for cite1 in range(pub,YearBoundary):
    for cite2 in range(cite1+1, YearBoundary):
        df1 = cur_pub_df[cur_pub_df['CiteYear'] == cite1]
        df2 = cur_pub_df[cur_pub_df['CiteYear'] == cite2]
        tempdf = df1.merge(df2, on="DocID", how="outer").fillna(0)
        temparray.append([pub, cite1, cite2, scipy.stats.pearsonr(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0], scipy.stats.kendalltau(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0]])
ax1 = fig.add_subplot(241)
ax1.set_title(str(pub+1996))
ax1.set_xlabel("Cited year 2")
ax1.set_ylabel("Cited year 1")
cmap=plt.cm.jet
cmap.set_under(color='white')     
figdf = pd.DataFrame(temparray, columns=["pubyear", "cite01", "cite02", "Pearson", "Kendaltau"])
heat1 = ax1.pcolor(figdf[["cite01", "cite02", "Pearson"]].pivot(index='cite01', columns='cite02', values='Pearson'), cmap = cmap)
heat1.set_clim(-1,1)
plt.colorbar(heat1)


pub = 4
cur_pub_df = countdata[(countdata['PubYear'] == pub) & (countdata['CiteCount'] != 0)]
temparray = []
for cite1 in range(pub,YearBoundary):
    for cite2 in range(cite1+1, YearBoundary):
        df1 = cur_pub_df[cur_pub_df['CiteYear'] == cite1]
        df2 = cur_pub_df[cur_pub_df['CiteYear'] == cite2]
        tempdf = df1.merge(df2, on="DocID", how="outer").fillna(0)
        temparray.append([pub, cite1, cite2, scipy.stats.pearsonr(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0], scipy.stats.kendalltau(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0]])   
ax2 = fig.add_subplot(242)
ax2.set_title(str(pub+1996))
ax2.set_xlabel("Cited year 2")
ax2.set_ylabel("Cited year 1")
cmap=plt.cm.jet
cmap.set_under(color='white')     
figdf = pd.DataFrame(temparray, columns=["pubyear", "cite01", "cite02", "Pearson", "Kendaltau"])
heat2 = ax2.pcolor(figdf[["cite01", "cite02", "Pearson"]].pivot(index='cite01', columns='cite02', values='Pearson'), cmap = cmap)
heat2.set_clim(-1,1)
plt.colorbar(heat2)


pub = 8
cur_pub_df = countdata[(countdata['PubYear'] == pub) & (countdata['CiteCount'] != 0)]
temparray = []
for cite1 in range(pub,YearBoundary):
    for cite2 in range(cite1+1, YearBoundary):
        df1 = cur_pub_df[cur_pub_df['CiteYear'] == cite1]
        df2 = cur_pub_df[cur_pub_df['CiteYear'] == cite2]
        tempdf = df1.merge(df2, on="DocID", how="outer").fillna(0)
        temparray.append([pub, cite1, cite2, scipy.stats.pearsonr(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0], scipy.stats.kendalltau(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0]])   
ax3 = fig.add_subplot(243)
ax3.set_title(str(pub+1996))
ax3.set_xlabel("Cited year 2")
ax3.set_ylabel("Cited year 1")
cmap=plt.cm.jet
cmap.set_under(color='white')     
figdf = pd.DataFrame(temparray, columns=["pubyear", "cite01", "cite02", "Pearson", "Kendaltau"])
heat3 = ax3.pcolor(figdf[["cite01", "cite02", "Pearson"]].pivot(index='cite01', columns='cite02', values='Pearson'), cmap = cmap)
heat3.set_clim(-1,1)
plt.colorbar(heat3)


pub = 12
cur_pub_df = countdata[(countdata['PubYear'] == pub) & (countdata['CiteCount'] != 0)]
temparray = []
for cite1 in range(pub,YearBoundary):
    for cite2 in range(cite1+1, YearBoundary):
        df1 = cur_pub_df[cur_pub_df['CiteYear'] == cite1]
        df2 = cur_pub_df[cur_pub_df['CiteYear'] == cite2]
        tempdf = df1.merge(df2, on="DocID", how="outer").fillna(0)
        temparray.append([pub, cite1, cite2, scipy.stats.pearsonr(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0], scipy.stats.kendalltau(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0]])   
ax4 = fig.add_subplot(244)
ax4.set_title(str(pub+1996))
ax4.set_xlabel("Cited year 2")
ax4.set_ylabel("Cited year 1")
cmap=plt.cm.jet
cmap.set_under(color='white')     
figdf = pd.DataFrame(temparray, columns=["pubyear", "cite01", "cite02", "Pearson", "Kendaltau"])
heat4 = ax4.pcolor(figdf[["cite01", "cite02", "Pearson"]].pivot(index='cite01', columns='cite02', values='Pearson'), cmap = cmap)
heat4.set_clim(-1,1)
plt.colorbar(heat4)


pub = 0
cur_pub_df = countdata_norm[(countdata_norm['PubYear'] == pub) & (countdata_norm['CiteCount'] != 0)]
temparray = []
for cite1 in range(pub,YearBoundary):
    for cite2 in range(cite1+1, YearBoundary):
        df1 = cur_pub_df[cur_pub_df['CiteYear'] == cite1]
        df2 = cur_pub_df[cur_pub_df['CiteYear'] == cite2]
        tempdf = df1.merge(df2, on="DocID", how="outer").fillna(0)
        temparray.append([pub, cite1, cite2, scipy.stats.pearsonr(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0], scipy.stats.kendalltau(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0]])
ax5 = fig.add_subplot(245)
ax5.set_title(str(pub+1996))
ax5.set_xlabel("Cited year 2")
ax5.set_ylabel("Cited year 1")
cmap=plt.cm.jet
cmap.set_under(color='white')     
figdf = pd.DataFrame(temparray, columns=["pubyear", "cite01", "cite02", "Pearson", "Kendaltau"])
heat5 = ax5.pcolor(figdf[["cite01", "cite02", "Pearson"]].pivot(index='cite01', columns='cite02', values='Pearson'), cmap = cmap)
heat5.set_clim(-1,1)
plt.colorbar(heat5)
    
    
pub = 4
cur_pub_df = countdata_norm[(countdata_norm['PubYear'] == pub) & (countdata_norm['CiteCount'] != 0)]
temparray = []
for cite1 in range(pub,YearBoundary):
    for cite2 in range(cite1+1, YearBoundary):
        df1 = cur_pub_df[cur_pub_df['CiteYear'] == cite1]
        df2 = cur_pub_df[cur_pub_df['CiteYear'] == cite2]
        tempdf = df1.merge(df2, on="DocID", how="outer").fillna(0)
        temparray.append([pub, cite1, cite2, scipy.stats.pearsonr(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0], scipy.stats.kendalltau(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0]])
ax6 = fig.add_subplot(246)
ax6.set_title(str(pub+1996))
ax6.set_xlabel("Cited year 2")
ax6.set_ylabel("Cited year 1")
cmap=plt.cm.jet
cmap.set_under(color='white')     
figdf = pd.DataFrame(temparray, columns=["pubyear", "cite01", "cite02", "Pearson", "Kendaltau"])
heat6 = ax6.pcolor(figdf[["cite01", "cite02", "Pearson"]].pivot(index='cite01', columns='cite02', values='Pearson'), cmap = cmap)
heat6.set_clim(-1,1)
plt.colorbar(heat6)
    
    
pub = 8
cur_pub_df = countdata_norm[(countdata_norm['PubYear'] == pub) & (countdata_norm['CiteCount'] != 0)]
temparray = []
for cite1 in range(pub,YearBoundary):
    for cite2 in range(cite1+1, YearBoundary):
        df1 = cur_pub_df[cur_pub_df['CiteYear'] == cite1]
        df2 = cur_pub_df[cur_pub_df['CiteYear'] == cite2]
        tempdf = df1.merge(df2, on="DocID", how="outer").fillna(0)
        temparray.append([pub, cite1, cite2, scipy.stats.pearsonr(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0], scipy.stats.kendalltau(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0]])
ax7 = fig.add_subplot(247)
ax7.set_title(str(pub+1996))
ax7.set_xlabel("Cited year 2")
ax7.set_ylabel("Cited year 1")
cmap=plt.cm.jet
cmap.set_under(color='white')     
figdf = pd.DataFrame(temparray, columns=["pubyear", "cite01", "cite02", "Pearson", "Kendaltau"])
heat7 = ax7.pcolor(figdf[["cite01", "cite02", "Pearson"]].pivot(index='cite01', columns='cite02', values='Pearson'), cmap = cmap)
heat7.set_clim(-1,1)
plt.colorbar(heat7)
    
    
pub = 12
cur_pub_df = countdata_norm[(countdata_norm['PubYear'] == pub) & (countdata_norm['CiteCount'] != 0)]
temparray = []
for cite1 in range(pub,YearBoundary):
    for cite2 in range(cite1+1, YearBoundary):
        df1 = cur_pub_df[cur_pub_df['CiteYear'] == cite1]
        df2 = cur_pub_df[cur_pub_df['CiteYear'] == cite2]
        tempdf = df1.merge(df2, on="DocID", how="outer").fillna(0)
        temparray.append([pub, cite1, cite2, scipy.stats.pearsonr(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0], scipy.stats.kendalltau(tempdf["CiteCount_x"], tempdf["CiteCount_y"])[0]])
ax8 = fig.add_subplot(248)
ax8.set_title(str(pub+1996))
ax8.set_xlabel("Cited year 2")
ax8.set_ylabel("Cited year 1")
cmap=plt.cm.jet
cmap.set_under(color='white')     
figdf = pd.DataFrame(temparray, columns=["pubyear", "cite01", "cite02", "Pearson", "Kendaltau"])
#print figdf[["cite01", "cite02", "Pearson"]].pivot(index='cite01', columns='cite02', values='Pearson')
heat8 = ax8.pcolor(figdf[["cite01", "cite02", "Pearson"]].pivot(index='cite01', columns='cite02', values='Pearson'), cmap = cmap)
heat8.set_clim(-1,1)
plt.colorbar(heat8)
    
labellist = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m"]
index = 0
for ax in fig.get_axes():
    if(index%2 == 0):
        ax.yaxis.label.set_fontsize(14)
        ax.xaxis.label.set_fontsize(14)       
        ax.tick_params(axis='x', labelsize=14)
        ax.tick_params(axis='y', labelsize=14)
        ax.text(-0.2, 1.1, labellist[index/2], fontsize=14, weight='bold', transform=ax.transAxes)  
        ax.set_xlabel("Cited year 2")
        ax.set_ylabel("Cited year 1")        
    if(index == 0):
        ax.text(0.05, 0.9, "Raw", fontsize=14, weight='bold', transform=ax.transAxes)
    if(index == 8):
        ax.text(0.05, 0.9, "Norm", fontsize=14, weight='bold', transform=ax.transAxes)          
    index += 1

    
plt.tight_layout()
plt.savefig("FIG006_PEARSON_R_CITATION.pdf")
plt.show()
```
