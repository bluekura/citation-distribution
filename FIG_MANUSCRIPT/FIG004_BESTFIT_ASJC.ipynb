```python
#%matplotlib inline  
#import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors as colors
import numpy as np
import csv
import glob
import os.path
import powerlaw
from collections import defaultdict
import pandas as pd

def discrete_cmap(N, base_cmap=None):
    base = plt.cm.get_cmap(base_cmap)
    color_list = base(np.linspace(0, 1, N))
    cmap_name = base.name + str(N)
    return base.from_list(cmap_name, color_list, N)

def cm2inch(value):
    return value/2.54
```

```python
nowdata = []

ASJC_SET = sorted(set(x.split("/")[-1].split("_")[1] for x in glob.glob("../DATA/distribution_compare/*.txt")))
ASJC_SET = [i for i in ASJC_SET if len(i) == 2]
ASJC_SET.sort(key = lambda first: int(first[:2]))

print len(ASJC_SET)
for ASJC_CODE in ASJC_SET:
    
    datafile = '../DATA/distribution_compare/ASJC_'+ ASJC_CODE + '.txt'
    if(not os.path.isfile(datafile)):
        continue
    count = defaultdict(float)
    with open(datafile, 'rb') as csvfile:
        reader = csv.reader(csvfile, delimiter = '\t')
        for row in reader:
            if(len(row) == 0):
                continue
            if(unicode(row[0]).isnumeric()):
                nowdata.append([int(row[0]), int(row[1]), float(row[3])])
                   
df = pd.DataFrame(nowdata, columns=['PubYear', "CiteYear", "BestFit"])
#print df[df['PubYear']==i]
nowdata2 = []
for i in range(1996,2018):
    for j in range(1996,2018):
        count = pd.value_counts(df[(df['PubYear']==i)&(df['CiteYear']==j)]['BestFit'].values)
        temparray = [i, j]
        for r in range(6):
            if r in count :
                now = count[r]
                temparray.append(now)
            elif(i>j):
                temparray.append(-1)
            else:
                temparray.append(0)
        nowdata2.append(temparray)
df2 = pd.DataFrame(nowdata2, columns=['PubYear', "CiteYear", 'PL', 'PLE', 'LN', 'LNP', 'EXP', 'StEX'])
#print df2

#plt.close()
#ASJC4

nowdata = []

ASJC_SET = sorted(set(x.split("/")[-1].split("_")[1] for x in glob.glob("../DATA/distribution_compare/*.txt")))
ASJC_SET = [i for i in ASJC_SET if len(i) == 4]
ASJC_SET.sort(key = lambda first: int(first[:4]))

max = 27
max2 = 309

#ASJC_CODE = "16"
print len(ASJC_SET)
for ASJC_CODE in ASJC_SET:
    datafile = './distribution_compare/ASJC_'+ ASJC_CODE + '.txt'
    if(not os.path.isfile(datafile)):
        continue
    count = defaultdict(float)
    with open(datafile, 'rb') as csvfile:
        reader = csv.reader(csvfile, delimiter = '\t')
        for row in reader:
            if(len(row) == 0):
                continue
            if(unicode(row[0]).isnumeric()):
                nowdata.append([int(row[0]), int(row[1]), float(row[3])])

                
df3 = pd.DataFrame(nowdata, columns=['PubYear', "CiteYear", "BestFit"])
#print df[df['PubYear']==i]
nowdata2 = []
for i in range(1996,2018):
    for j in range(1996,2018):
        count = pd.value_counts(df3[(df3['PubYear']==i)&(df3['CiteYear']==j)]['BestFit'].values)
        temparray = [i, j]
        for r in range(6):
            if r in count :
                now = count[r]
                temparray.append(now)
            elif(i>j):
                temparray.append(-1)
            else:
                temparray.append(0)
        nowdata2.append(temparray)

df4 = pd.DataFrame(nowdata2, columns=['PubYear', "CiteYear", 'PL', 'PLE', 'LN', 'LNP', 'EXP', 'StEX'])

#print df2
#print df4

fig = plt.figure(figsize =(cm2inch(19)*2,cm2inch(11)*2))
cmap=plt.cm.jet
cmap.set_under(color='white') 
ax1 = fig.add_subplot(231)
ax1.set_title("PL")
ax1.set_xlabel("PubYear")
ax1.set_ylabel("CiteYear")
ax1.text(0.05, 0.9, "Subject areas", fontsize=14, weight='bold', transform=ax1.transAxes)
heat1 = ax1.pcolor(df2.pivot(index='PubYear', columns='CiteYear', values='PL'), cmap = cmap)
heat1.set_clim(0,max)
plt.colorbar(heat1)

ax2 = fig.add_subplot(232)
ax2.set_title("PLE")
ax2.set_xlabel("PubYear")
ax2.set_ylabel("CiteYear")
heat2 = ax2.pcolor(df2.pivot(index='PubYear', columns='CiteYear', values='PLE'), cmap = cmap)
heat2.set_clim(0,max)
plt.colorbar(heat2)

ax3 = fig.add_subplot(233)
ax3.set_title("LN")
ax3.set_xlabel("PubYear")
ax3.set_ylabel("CiteYear")
heat3 = ax3.pcolor(df2.pivot(index='PubYear', columns='CiteYear', values='LN'), cmap = cmap)
heat3.set_clim(0,max)
plt.colorbar(heat3)

ax4 = fig.add_subplot(234)
ax4.set_title("PL")
ax4.set_xlabel("PubYear")
ax4.set_ylabel("CiteYear")
ax4.text(0.05, 0.9, "Subject categories", fontsize=14, weight='bold', transform=ax4.transAxes)
heat4 = ax4.pcolor(df4.pivot(index='PubYear', columns='CiteYear', values='PL'), cmap = cmap)
heat4.set_clim(0,max2)
plt.colorbar(heat4)

ax5 = fig.add_subplot(235)
ax5.set_title("PLE")
ax5.set_xlabel("PubYear")
ax5.set_ylabel("CiteYear")
heat5 = ax5.pcolor(df4.pivot(index='PubYear', columns='CiteYear', values='PLE'), cmap = cmap)
heat5.set_clim(0,max2)
plt.colorbar(heat5)

ax6 = fig.add_subplot(236)
ax6.set_title("LN")
ax6.set_xlabel("PubYear")
ax6.set_ylabel("CiteYear")
heat6 = ax6.pcolor(df4.pivot(index='PubYear', columns='CiteYear', values='LN'), cmap = cmap)
heat6.set_clim(0,max2)
plt.colorbar(heat6)

labellist = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m"]
index = 0
for ax in fig.get_axes():
    if(index%2 == 0):
        ax.yaxis.label.set_fontsize(14)
        ax.xaxis.label.set_fontsize(14)
        ax.set_ylabel("Published Year")
        ax.set_xlabel("Cited Year")        
        ax.tick_params(axis='x', labelsize=14)
        ax.tick_params(axis='y', labelsize=14)       
        xlabels = [i for i in range(1996, 2018, 4)]
        ylabels =  [i for i in range(1996, 2018, 4)]
        ax.set_xticks([i for i in range(0, 21, 4)])
        ax.set_yticks([i for i in range(0, 21, 4)])
        ax.set_xticklabels(xlabels)
        ax.set_yticklabels(ylabels)        
        ax.text(-0.2, 1, labellist[index/2], fontsize=14, weight='bold', transform=ax.transAxes)  
    index += 1


plt.tight_layout()
plt.savefig('FIG004_BEST_DIST_RAW.pdf')
plt.show()
```
