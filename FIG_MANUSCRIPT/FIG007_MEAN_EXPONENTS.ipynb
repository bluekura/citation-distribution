```python
#%matplotlib inline
import gc
import pandas as pd
import numpy as np
import scipy.stats
import matplotlib.pyplot as plt
import matplotlib.colors as colors

def cm2inch(value):
    return value/2.54
```

```python
file_asjc2 = "../DATA/asjc_2_pdf_exponents.tsv"
df_asjc2 = pd.read_csv(file_asjc2, delimiter ="\t", index_col=0)
```

```python
file_asjc2 = "../DATA/asjc_2_pdf_exponents.tsv"
df_asjc2 = pd.read_csv(file_asjc2, delimiter ="\t", index_col=0)
plt.figure()
df_asjc2[['Exponent_Unnorm','Exponent_Norm']].plot()
plt.fill_between(df_asjc2.index, df_asjc2['Exponent_Unnorm'] + df_asjc2['Std.Dev_Unnorm'], df_asjc2['Exponent_Unnorm'] - df_asjc2['Std.Dev_Unnorm'], color='blue', alpha=0.2)
plt.fill_between(df_asjc2.index, df_asjc2['Exponent_Norm'] + df_asjc2['Std.Dev_Norm'], df_asjc2['Exponent_Norm'] - df_asjc2['Std.Dev_Norm'], color='orange', alpha=0.2)
plt.legend(labels=["$C_y(a)$", "$C^{*}_y(a)$"])
```

```python
file_asjc4 = "../DATA/asjc_4_pdf_exponents.tsv"
df_asjc4 = pd.read_csv(file_asjc4, delimiter ="\t", index_col=0)
plt.figure()
df_asjc4[['Exponent_Unnorm','Exponent_Norm']].plot()
plt.fill_between(df_asjc4.index, df_asjc2['Exponent_Unnorm'] + df_asjc4['Std.Dev_Unnorm'], df_asjc4['Exponent_Unnorm'] - df_asjc4['Std.Dev_Unnorm'], color='blue', alpha=0.2)
plt.fill_between(df_asjc4.index, df_asjc2['Exponent_Norm'] + df_asjc4['Std.Dev_Norm'], df_asjc4['Exponent_Norm'] - df_asjc4['Std.Dev_Norm'], color='orange', alpha=0.2)
plt.legend(labels=["$C_y(a)$", "$C^{*}_y(a)$"])
```

```python
fig = plt.figure(figsize =(cm2inch(14)*2,cm2inch(6)*2))

ax1 = fig.add_subplot(121)
df_asjc2[['Exponent_Unnorm','Exponent_Norm']].plot(ax=ax1)
ax1.fill_between(df_asjc2.index, df_asjc2['Exponent_Unnorm'] + df_asjc2['Std.Dev_Unnorm'], df_asjc2['Exponent_Unnorm'] - df_asjc2['Std.Dev_Unnorm'], color='blue', alpha=0.2)
ax1.fill_between(df_asjc2.index, df_asjc2['Exponent_Norm'] + df_asjc2['Std.Dev_Norm'], df_asjc2['Exponent_Norm'] - df_asjc2['Std.Dev_Norm'], color='orange', alpha=0.2)
plt.legend(labels=["$C_y(a)$", "$C^{*}_y(a)$"], fontsize = 14)


ax2 = fig.add_subplot(122)
df_asjc4[['Exponent_Unnorm','Exponent_Norm']].plot(ax=ax2)
plt.fill_between(df_asjc4.index, df_asjc2['Exponent_Unnorm'] + df_asjc4['Std.Dev_Unnorm'], df_asjc4['Exponent_Unnorm'] - df_asjc4['Std.Dev_Unnorm'], color='blue', alpha=0.2)
plt.fill_between(df_asjc4.index, df_asjc2['Exponent_Norm'] + df_asjc4['Std.Dev_Norm'], df_asjc4['Exponent_Norm'] - df_asjc4['Std.Dev_Norm'], color='orange', alpha=0.2)
plt.legend(labels=["$C_y(a)$", "$C^{*}_y(a)$"], fontsize = 14)
ax2.set_xlabel('Avrage number of citations', fontsize = 14)
ax2.set_ylabel('PDF', fontsize = 14)


labellist = ["a", "b", "c", "d", "e", "f", "g", "h", "i", "j", "k", "l", "m"]
index = 0

for ax in fig.get_axes():
    ax.yaxis.label.set_fontsize(14)
    ax.xaxis.label.set_fontsize(14)       
    ax.tick_params(axis='x', labelsize=14)
    ax.tick_params(axis='y', labelsize=14) 
    ax.text(-0.12, 1, labellist[index], fontsize=14, weight='bold', transform=ax.transAxes)  
    ax.set_xlabel('Published year - Cited year', fontsize = 14)
    ax.set_ylabel('Power-law exponent', fontsize = 14)
    if(index == 0):
        ax.set_yticks([2,4,6,8,10])      
        ax.set_title("27 Subject areas", fontsize=14)
    if(index == 1):
        ax.set_yticks([0,2,4,6,8,10,12])
        ax.set_title("309 Subject categories", fontsize=14)
    index += 1
    
plt.tight_layout()
plt.savefig("FIG007_AVG_EXPONENT.pdf")

plt.show()
```
